```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.imdb.com/search/title/?groups=top_100&view=advanced')
page2=requests.get('https://www.imdb.com/search/title/?groups=top_100&start=51&ref_=adv_nxt')
print(page,page2)
```

```python
soup=BeautifulSoup (page.content)
soup2=BeautifulSoup(page2.content)
```

```python
movie_title1=soup.find(class_="lister-item-header")
movie_title2=soup2.find(class_="lister-item-header")
titles=movie_title1.text.split()[1:8],movie_title2.text.split()[1:7]
print(titles)
```

```python
title=[]
for i in soup.find_all(class_="lister-item-header"):
    title.append(i.text.split()[1])
for h in soup2.find_all(class_="lister-item-header"):
    title.append(h.text.split()[1])

print(title)
```

```python
year1=soup.find(class_="lister-item-year text-muted unbold")
year2=soup2.find(class_="lister-item-year text-muted unbold")
years=year1.text,year2.text
print(years)
```

```python
year=[]
for i in soup.find_all(class_="lister-item-year text-muted unbold"):
    year.append(i.text)
for h in soup2.find_all(class_="lister-item-year text-muted unbold"):
    year.append(h.text)

print(year)
```

```python
page=requests.get('https://www.imdb.com/search/title/?groups=top_100&view=advanced')
page2=requests.get('https://www.imdb.com/search/title/?groups=top_100&start=51&ref_=adv_prv')
print(page,page2)
```

```python
movie_rating1=soup.find(class_="inline-block ratings-imdb-rating")
print(movie_rating1.text)
movie_rating2=soup2.find(class_="inline-block ratings-imdb-rating")
print(movie_rating1.text,movie_rating2.text)
```

```python
rating=[]
for i in soup.find_all(class_="inline-block ratings-imdb-rating"):
    rating.append(i.text)
for h in soup2.find_all(class_="inline-block ratings-imdb-rating"):
    rating.append(h.text)

print(rating,sep='\n')
```

```python
print (len(title),len(rating),len (year))
```

```python
import pandas as pd
df=pd.DataFrame({'titles':title,'rating':rating,'year':year})
print(df)
```

#Q4 4) Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp . 

```python
page=requests.get("https://meesho.com/bags-ladies/pl/p7vbp")
soup=BeautifulSoup(page.content)
product_name=soup.find(class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS")
print(product_name.text)
```

```python
Name=[]
for k in soup.find_all(class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    Name.append(k.text)
    
print(Name)
```

```python
product_price=soup.find(class_="Text__StyledText-sc-oo0kvp-0 hiHdyy")
print(product_price.text.replace('₹',''))
```

```python
Price=[]
for p in soup.find_all(class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    Price.append(p.text.replace('₹',''))
print(Price)
```

```python
product_discount=soup.find(class_="Text__StyledText-sc-oo0kvp-0 fCJVtz NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm")
print(product_discount.text.split()[0].replace('₹',''))
```

```python
Discount=[]
for d in soup.find_all(class_="Text__StyledText-sc-oo0kvp-0 fCJVtz NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm"):
    Discount.append(d.text.split()[0].replace('₹',''))
print(Discount)
 

```

```python
print(len(Name),len(Price),len(Discount))
```

```python
import pandas as pd
meesho=pd.DataFrame({'Name':Name,'Price':Price,'Discount':Discount})
print (meesho)
```
